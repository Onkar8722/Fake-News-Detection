```python
import copy
from nltk.corpus import stopwords
import string
```

```python
# removing punctuations from title
import string

def text_process(title):
    nop = [char for char in title if char not in string.punctuation]
    nop = ''.join(nop)
    return [word for word in nop.split() if word in word.lower() not in nltk.corpus.stopwords.words('english')]
```

```python
# importing libraries for classification model
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#Naive model with hyper parameters
piplineTitle = Pipeline([
    ('bow', CountVectorizer(analyzer=text_process)),
    ('tfidf', TfidfTransformer()),
    ('classifier', MultinomialNB()),
])
```

```python
# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(final_data['title'], final_data['label'], test_size=0.2, random_state=123)
```

```python
import nltk
nltk.download('stopwords')
import string
#create a function for analyzing the weights and keywords from data
def text_process(title):
    nop = [char for char in title if char not in string.punctuation]
    nop = ''.join(nop)
    return [word for word in nop.split() if word.lower() not in nltk.corpus.stopwords.words('english')]
```

```python
#creating a model pipeline to feed the data to model
import nltk
print(piplineTitle.fit(X_train, y_train))
```

```python
#creating predict function to predict the output of test dataset
y_pred = piplineTitle.fit(X_train, y_train).predict(X_test)
y_pred
```

```python
# classification report
clf_report = classification_report(y_test, y_pred)
print('Classification_Report',clf_report)
```

```python
cnf_matrix = confusion_matrix(y_test, y_pred)
print('Cnfusion Matrix',cnf_matrix)
```

```python
# Visualizing confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (10,7))
sns.heatmap(cnf_matrix, annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```

```python
# Function to predict news category
def predict_news(text):
    prediction = piplineTitle.predict([text])[0]
    return prediction
```
